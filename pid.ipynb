```python
import matplotlib.pyplot as plt


def pid(setpoint , previous_error , pv , kp , ki , kd , ie , dt):
    error = setpoint  - pv
    ie += error*dt
    de = (error - previous_error)/dt
    c_o = kp*error + ki*ie + kd*de
    return c_o , ie , error


def main():
    setpoint = 100
    previous_error = 0
    pv = 0
    kp = 1.5
    ki = 0.1
    kd = 0.05
    dt = 0.1
    ie = 0                                                                                                                                                 
    time = []
    c_values = []
    pv_values  = []

    for i in range(100):
        
        c_o , ie , error = pid(setpoint , previous_error , pv , kp , ki , kd , ie , dt)
        previous_error = error
        pv += c_o*dt
        
        time.append(i*dt)
        c_values.append(c_o)
        pv_values.append(pv)

    plt.plot(time , [setpoint]*100 , linestyle = '--' )
    plt.plot(time , pv_values, label = 'PV')
    plt.plot(time, c_values , label = 'Control Output')
    plt.legend()
    plt.show()


if __name__ == "__main__":
        main()
     

        

    
    
```
